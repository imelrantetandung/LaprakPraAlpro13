```python
#SOAL NO.1
n = int(input('Masukkan nilai n: '))

data_aplikasi = {}
for i in range(n):
    kategori = input('Masukkan nama kategori:')
    print('Masukkan 5 nama aplikasi di kategori', kategori)

    aplikasi = []
    for j in range(5):
        nama_aplikasi = input('Nama aplikasi: ')
        aplikasi.append(nama_aplikasi)

    data_aplikasi[kategori] = aplikasi
print(data_aplikasi)

daftar_aplikasi_list = []
for aplikasi in data_aplikasi.values():
    daftar_aplikasi_list.append(set(aplikasi))
print(daftar_aplikasi_list)

hasil = daftar_aplikasi_list[0]
for i in range(1, len(daftar_aplikasi_list)):
    hasil = hasil.intersection(daftar_aplikasi_list[i])
print(hasil)
print()
#Daftar aplikasi yang hanya mucul di 1 kategori
hasil2 = daftar_aplikasi_list[0]
for i in range(1, len(daftar_aplikasi_list)):
    hasil2 = hasil2.symmetric_difference(daftar_aplikasi_list[i])
print('Aplikasi yang hanya muncul di 1 kategori: ',hasil2)
print()

#untuk n > 2, daftar aplikasi yang mucul tepat di 2 kategori
frekuensi = dict()
for i in daftar_aplikasi_list:
    for j in i:
        if j not in frekuensi:
            frekuensi[j] = 1
        else:
            frekuensi[j] += 1

hasil3 = set()
for i in frekuensi.keys():
    if frekuensi[i] == 2:
        hasil3.add(i)

print('Aplikasi yang muncul tepat di 2 kategori:',hasil3)
```

```python
#SOAL NO.2
list1 = [1, 2, 3, 4, 5, 5, 2]
print("List sebelum konversi:", list1)

set_menjadi_list = set(list1)
print("Set setelah konversi dari List:", set_menjadi_list)
print()

set1 = {1, 2, 3, 4, 5}
print("Set sebelum konversi:", set1)

list_ke_set = list(set1)
print("List setelah konversi dari Set:", list_ke_set)
print()

tuple1 = (1, 2, 3, 4, 5, 5, 2)
print("Tuple sebelum konversi:", tuple1)

set_menjadi_tuple = set(tuple1)
print("Set setelah konversi dari Tuple:", set_menjadi_tuple)
print()

print("Set sebelum konversi:", set1)

tuple_menjadi_set = tuple(set1)
print("Tuple setelah konversi dari Set:", tuple_menjadi_set)
print()
```

```python
#SOAL NO.3
file1= input('Masukkan file pertama: ')
file2= input('Masukkan file kedua: ')

try:
    file_pertama = open(file1)
except:
    print('File cannot be opened: ', file1)
    exit()

try:
    file_kedua = open(file2)
except:
    print('File cannot be opened: ', file2)
    exit()

daftar_kata = []
for i in file_pertama:
    i = i.strip()
    i = i.split()
    for j in i:
        j = j.lower()
        daftar_kata.append(j)

daftar_kata2 = []
for i in file_kedua:
    i = i.strip()
    i = i.split()
    for j in i:
        j = j.lower()
        daftar_kata2.append(j)

kata_sama = set(daftar_kata) & set(daftar_kata2)
print('Daftar kata yang sama: ')
for i in kata_sama:
    print(i)

file_pertama.close()
file_kedua.close()
```
